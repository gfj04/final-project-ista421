# Final Project Part C: Project Check-In and Outline

## 1. Introduction & Research Question
The goal of this project is to predict vehicle fuel efficiency (MPG) using machine learning models.

Research Question: How accurately can we predict MPG using vehicle characteristics such as weight, horsepower, and displacement?


## 2. Significance
Fuel efficiency is important for both cost savings and environmental impact. Understanding what affects MPG can help improve vehicle design and reduce fuel consumption. This project uses machine learning to analyze real-world data and identify the most important factors affecting MPG.

## 3. Dataset
The dataset used is the Auto MPG dataset.

Target variable: MPG  
Features used: Weight, Horsepower, Displacement  

These features were selected because they are directly related to vehicle performance and fuel efficiency.

## 4. Data Preprocessing
Removed missing values;
Selected relevant features;  
Split the dataset into training and testing sets (80/20 split)


## 5. Models Used

### Model 1: Simple Linear Regression

Feature: Weight only  
Purpose: baseline model  
Reason: simple and easy to interpret relationship between weight and MPG  

This model serves as a baseline to understand the relationship between weight and MPG. 

### Model 2: Multiple Linear Regression
 Features: Weight, Horsepower, Displacement  
 Purpose: improve prediction accuracy  
 Reason: MPG depends on multiple factors, not just one variable  

Linear regression was chosen because it is simple and interpretable. The multiple regression model helps capture more complex relationships between features and MPG.

Future work could include polynomial regression to capture non-linear relationships.


## 6. Evaluation Metrics
The models were evaluated using:

 Mean Squared Error (MSE)  
 Mean Absolute Error (MAE)  
 R² Score  

These metrics help evaluate how well the model predicts MPG and compare performance between models.

The results showed that using multiple features did not significantly improve performance compared to using only weight, indicating that weight is the most important predictor of MPG.
